```python
!pip install transformers
from transformers import pipeline
emotion_model=pipeline('text-classification',model='j-hartmann/emotion-english-distilroberta-base')
```

```python
def identify_emotion(text):
    result=emotion_model(text)
    return result
```

```python
user_input=input('Enter a sentence: ')
emotion=identify_emotion(user_input)
print("Detected Emotion:",emotion[0]['label'],"with a confidence of",emotion[0]['score'])
```

```python
import pytesseract
import cv2
from PIL import Image
import matplotlib.pyplot as plt
from transformers import pipeline
from google.colab import files
import numpy as np
```

```python
#upload image
uploaded = files.upload()
image_path = list(uploaded.keys())[0]
```

```python
#Load and display the image
img = cv2.imread(image_path)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
#use pytesseract to extract text
extracted_text = pytesseract.image_to_string(img_rgb)
```

```python
#Display the image and extracted text
plt.imshow(img_rgb)
plt.axis('off')
plt.show()
print("Extracted Text:\n",extracted_text)
#perform sentiment analysis using Hugging Face pipeline
sentiment_analyzer = pipeline('sentiment-analysis')
sentiment_result = sentiment_analyzer(extracted_text)
label=sentiment_result[0]['label']
score=sentiment_result[0]['score']
if label=='POSITIVE':
  emoji="😊"  # Assign a positive emoji
elif label=='NEGATIVE':
  emoji="😔"  # Assign a negative emoji
else:
  emoji="😐"  # Assign a neutral emoji

print(f"Sentiment: {label} {emoji} (Score: {score})")
```
