```python
population_data.drop(['Indicator Name','Country Name'],axis=1,inplace=True)
```

```python
population_data_cleaned = population_data.loc[ : , ~population_data.columns.str.contains('^Unnamed')]

population_data_filled = population_data_cleaned.fillna(method="ffill")

# Check if there are any remaining missing values
print("Any remaining missing values after forward fill:", population_data_filled.isna().sum().any())

# If there are still missing values, let's inspect them
if population_data_filled.isna().sum().any():
    print("Columns with missing values:")
    print(population_data_filled.isna().sum())

# Display the first few rows of the filled data
print(population_data_filled.head())
```

```python
population_data_filled.columns
```

```python
# Check unique values in the 'Indicator Code' column
unique_indicator_codes = population_data_filled['Indicator Code'].unique()
print("Unique 'Indicator Code' values:", unique_indicator_codes)

# Assuming 'SP.POP.TOTL' is the correct code, we will filter the data using this code.
# Adjust this code based on the actual values found.
correct_indicator_code = 'SP.POP.TOTL'  # Adjust if needed after inspecting the unique values

# Filter the data to include only rows where 'Indicator Code' is the correct one
total_population_data = population_data_filled[population_data_filled['Indicator Code'] == correct_indicator_code]

# Display the first few rows of the filtered data
print(total_population_data.head())

# Sort data based on the total population for 2022
total_population_sorted = total_population_data.sort_values(by="2022", ascending=False)

# Get the top ten countries with the highest total population for 2022
total_top_ten_countries = total_population_sorted.head(10)
print("Top ten countries of total population\n")
print(total_top_ten_countries[['Country Code']])
```

```python
import matplotlib.pyplot as plt
# Plotting population data for the top ten countries over the years
cols = [str(year) for year in range(1960, 2024)]  # Adjusting years dynamically

for i in cols:
    fig = plt.figure(figsize=(10, 6))
    plt.bar(total_top_ten_countries['Country Code'], total_top_ten_countries[i], color='#B22222')
    plt.xlabel("Country Name")
    plt.ylabel("Population")
    plt.title(f"Total Population in {i}")
    plt.xticks(rotation=45)
    plt.show()
```

```python
# Create the bar plot
plt.figure(figsize=(15, 6))
plt.subplot(2,2,1)
sns.barplot(x="2022", y="Country Code", data=total_top_ten_countries, palette="coolwarm")
plt.title("Top Ten Countries of Total Population (2022)",fontsize=10)
plt.xlabel("Total Population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()

# Create the bar plot
plt.figure(figsize=(15, 6))
plt.subplot(2,2,2)
sns.barplot(x="2016", y="Country Code", data=total_top_ten_countries, palette="coolwarm")
plt.title("Top Ten Countries of Total Population (2016)",fontsize=10)
plt.xlabel("Total Population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()
```
